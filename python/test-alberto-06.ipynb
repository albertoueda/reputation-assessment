```python
import sys
sys.path.append('../../../../../ufmg-latin/pscore/rfs/')

import rfslib.datasets
dataset = rfslib.datasets.DBLPDataset('../../../../../pscore/datasets/',nauthors=10000)
dataset.prepare()

import rfslib.seriestrie
trie_authors = rfslib.seriestrie.SeriesTrie(dataset.dfauthors.Akey)
trie_venues  = rfslib.seriestrie.SeriesTrie(
                    dataset.dfvenues.Vkey.map(lambda x: x.replace('/',' ')))
trie_groups  = rfslib.seriestrie.SeriesTrie(
                    dataset.dfgroups.Name.map(lambda x: x.replace('-',' ')))

import rfslib.rankmodels
import pandas as pd
```

```python
dataset.df
```

```python
refids_v = trie_venues.search('conf sigir')

# Rank authors

authorrank  = dataset.dfauthors.copy()

year = [1940,2015]
authorscore = rfslib.rankmodels.RankModelVACount(dataset,refids_v,year=year).rank() # Venue -> Authors

field = 'Score_Test'
rfslib.rankmodels.addscore(authorrank,authorscore,field)
```
