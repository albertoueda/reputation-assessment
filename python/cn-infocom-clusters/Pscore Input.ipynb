```python
def print_top(l):
    k = 1
    for i in l:
        print('%d,' % i, end="")
        if k % 15 == 0: print("")
        k += 1
            
    print("\n")

def print_input(venue_name):    
    top = pd.read_csv(mypath + venue_name + '/' + 'gephi_top_' + venue_name + '.csv')

    print ("=============== ", venue_name, " ===============")
    print ("50 TOP (50 autores):")
    print_top(top[top['top'] == 1]['id'].tolist())

    top_center = top[top['filter_ref_set_10_center'] == 1]['id'].tolist()
    print ("10 TOP CENTER (", len(top_center), "):")
    print_top(top_center)

    top_largest = top[top['filter_ref_set_5_largest'] == 1]['id'].tolist()
    print ("5 LARGEST CLUSTERS (", len(top_largest), "):")
    print_top(top_largest)
    
```
