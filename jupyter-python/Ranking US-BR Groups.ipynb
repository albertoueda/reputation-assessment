```python
%pylab inline

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
input_dir = 'input/marlon/subarea group rank/'
subareas = ['Information retrieval', 'Database', 'Computer network', 
            'Parallel computing', 'Computer graphics (images)', 'Data mining', 'Machine learning']
nprograms = 20

for subarea in subareas:
    ranking = pd.read_csv(input_dir + subarea + '.csv', index_col=0, nrows=151)
    ranking = ranking.sort_values(by=subarea + ' year', ascending=False)
    ranking['logScore'] = log(ranking[subarea + ' year']) #ranking[subarea + ' year'] 

    ax = ranking.head(nprograms).plot(kind='bar', title=subarea, y='logScore', legend=False, width=0.4, figsize=(10,6))
    ax.set_xticklabels(ranking['Name'].values, rotation=45, ha='right')
    ax.set_ylabel('Log of Weighted-P-score')

    plt.tight_layout()
    savefig('output/2017spire/ranking-of-groups-' + subarea + '.eps')
```
