```python
%pylab inline

import pandas as pd
import numpy as np

from unidecode import unidecode
```

```python
dblp = pd.read_csv('../cnpq-authors-vkeys.csv')
mas = pd.read_csv('../mas-venue-subdomain.csv')

# Pre-processing tables
mas.rename(columns={'VenueOldID':'MASID', 'FullName':'MASName', 'Type':'MASType', 'SubDomainName':'MASSubDomainName'}, inplace=True)
mas = mas.ix[:, ['MASID', 'MASType', 'ShortName', 'MASName', 'DomainName', 'MASSubDomainName', 'DomainID', 'SubDomainID']]

mas['ShortName'] = mas[mas['ShortName'].notnull()]['ShortName'].apply(lambda x: unidecode(str(x).decode('utf-8)')).lower())
mas['MASName'] = mas['MASName'].apply(lambda x: unidecode(str(x).decode('utf-8)')).lower())

dblp.rename(columns={'VKey':'DBLPVenueKey', 'NameAZList':'DBLPName'}, inplace=True)
dblp['ShortNameAZList'] = dblp[dblp['ShortNameAZList'].notnull()]['ShortNameAZList'].apply(lambda x: unidecode(str(x).decode('utf-8)')).lower())
dblp['DBLPName'] = dblp['DBLPName'].apply(lambda x: unidecode(str(x).decode('utf-8)')).lower())

dblp['DBLPType'] = dblp['DBLPVenueKey'].apply(lambda x: 'C' if x.startswith('conf/') else 'J')
dblp['DBLPMinVKey'] = dblp['DBLPVenueKey'].apply(lambda x: unidecode(str(x).replace("conf/", "")
                                                                     .replace("journals/", "")
                                                                     .decode('utf-8)')).lower())

# NaN Workaround
# mas = mas[mas['ShortName'].notnull()]

# Merging tables
result = pd.merge(dblp, mas, 
                  left_on=['DBLPType', 'DBLPName'], right_on=['MASType', 'MASName'], how='left')

# Sort
result.sort(['MASType', 'DBLPVenueKey'], inplace=True)
result.drop_duplicates('DBLPVenueKey', inplace=True)
result.reset_index(inplace=True)

# Writing result files
result = result.ix[:, ['DBLPVenueKey', 'MASID', 'MASType','DBLPType', 'MASSubDomainName', 'ShortName', 'ShortNameAZList', 'DBLPName', 'DomainName', 'DomainID', 'SubDomainID']]
result.to_csv('../conciliated_venues_FULL.csv')

mini_result = result.ix[:, ['DBLPVenueKey', 'MASID', 'MASSubDomainName']]
mini_result.to_csv('../conciliated_venues.csv')
```

```python
# H-Index File
h = pd.read_csv('../cnpq/prod-levels.csv')
h.rename(columns={'Name':'Author', 'H-index':'HIndex'}, inplace=True)
h = h.ix[:, ['Author', 'HIndex']]
h['HIndex'] = h[h['HIndex'].notnull()]['HIndex'].apply(lambda x: str(int(x)))
h.to_csv('../author-hindex.csv')
```
