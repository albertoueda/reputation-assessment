```python
def int2hex(x):
    return ('%8s' % hex(x).upper()[2:]).replace(' ','0')

def hex2int(x):
    #try:
    return int(x,16)
    #except:
    #    print 'ERROR',x

def mag_read_df(filename,columns,keys,drop_keys=False,nrows=None):
    df = pd.read_csv(filename,sep='\t',names=columns,nrows=nrows)
    for k,v in keys:
        df[v] = df[k].map(hex2int)
        if drop_keys:
            del df[k]
    if len(keys) > 0:
        df.set_index(keys[0][1],inplace=True)
    return df
```

## Score

```python
YEAR = 2015

#df = dfpaperinst2.copy()
df = dfpaperinst2.loc[dfpaperinst2.Pyear==YEAR].copy()

# number of paper authors
df['PAcount' ] = df.groupby(level=0)['Akey'].transform('count')
df['PAIcount'] = df.groupby(['Pkey','Akey'])['Ikey'].transform('count')
df['Iscore'] = (1.0 / df['PAcount']) / df['PAIcount']
df
```

```python
df.groupby('Inorm').sum().sort('Iscore',ascending=False).loc[:,['PAIcount','Iscore']]
```
